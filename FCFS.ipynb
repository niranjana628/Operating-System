### First Come First Serve 

```python
def fcfs(n,pid,bt,at):
    for m in range(n):
        if m==n-1:
            break
        else:
            if at[m]>at[m+1]:
                temp=bt[m]
                bt[m]=bt[m+1]
                bt[m+1]=temp
                
                temp=at[m]
                at[m]=at[m+1]
                at[m+1]=temp
                
                temp=pid[m]
                pid[m]=pid[m+1]
                pid[m+1]=temp
    ct=[]
    ctime=0
    ttime=0
    tat=[]
    wtime=0
    wt=[]
    for i in range(n):
        if at[i]==0:
                ctime+=bt[i]
                ct.append(ctime)
                ttime=ctime-at[i]
                tat.append(ttime)
                wtime=ttime-bt[i]
                wt.append(wtime)
        else:
            if at[i]>ct[i-1]:
                ctime+=(bt[i]+(at[i]-ct[i-1]))
                ct.append(ctime)
                ttime=ctime-at[i]
                tat.append(ttime)
                wtime=ttime-bt[i]
                wt.append(wtime)
            else:
                ctime+=bt[i]
                ct.append(ctime)
                ttime=ctime-at[i]
                tat.append(ttime)
                wtime=ttime-bt[i]
                wt.append(wtime)
                
    
    print("pid arrival burst completion t.a time waiting")            
    for x in range(n):
        
        print(pid[x],'   ',at[x],'   ',bt[x],'     ',ct[x],'    ',tat[x],'    ',wt[x])
        
n=int(input('enter number of processes='))
pid=[]
at=[]
bt=[]
for i in range (n):
    pid.append(i+1)
    print('Enter the arrival time of process',(i+1),'=')
    a=int(input())
    at.append(a)
    print('Enter the burst time of process',(i+1),'=')
    b=int(input())
    bt.append(b)
print("pid   arrival  burst ")            
for x in range(n):
        
    print(pid[x],'  ',at[x],'  ',bt[x],'  ')    

sjf(n,pid,bt,at)
```
